```python
# Preparing data table for analysis
ratings_pivot = pd.pivot_table(data=books1, values='book_rating', index='userid', columns='book_title')
ratings_pivot
```

```python
correlation_matrix  = ratings_pivot.corr(method='pearson')
correlation_matrix
```

```python
#Impute those NaNs with 0 values
ratings_pivot.fillna(0, inplace=True)
```

```python
ratings_pivot
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(ratings_pivot.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books1.userid.unique()
user_sim_df.columns = books1.userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
